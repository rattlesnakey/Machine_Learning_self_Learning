```python
import requests
import json
import re
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url="https://vip.stock.finance.sina.com.cn/q/view/vFutures_History.php?jys=shfe&pz=AU&hy=AU0&breed=AU0&type=inner&start=2019-04-05&end=2020-05-04"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
```

```python
res=requests.get(url,headers=headers,timeout=None)
soup=BeautifulSoup(res.text,'lxml')
```

```python
def crawl(soup):
    s=soup.find_all("table")
    s=s[3]
    s=s.find_all("td")
    cnt=0
    begin=5
    s=s[6:]
    cnt=0
    all_dict={}
    temp=[]
    # for i in s:
    #     print(i.div)
    for i in s:
        if cnt%6==0:
            all_dict[i.div.string]=[]
            nowday=i.div.string
            cnt+=1
            continue
        all_dict[nowday].append(i.div.string)
        cnt+=1
    return all_dict
all_dict=crawl(soup)
```

```python
df1=pd.DataFrame(all_dict)
```

```python
df1.rename(index={0:'收盘价',1:'开盘价',2:'最高价',3:'最低价',4:'成交量'},inplace=True)
print(df1)
```

```python
url="https://vip.stock.finance.sina.com.cn/q//view/vFutures_History.php?page={}&breed=AG0&start=2019-04-05&end=2020-05-05&jys=shfe&pz=AG&hy=AG0&type=inner&name=大豆1109"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'}
x=list(range(1,8))
df=pd.DataFrame()
for i in x:
    tempurl=url.format(i)
    res=requests.get(tempurl,headers=headers,timeout=None)
    soup=BeautifulSoup(res.text,'lxml')
    tempdict=crawl(soup)
    tempdf=pd.DataFrame(tempdict)
    tempdf.rename(index={0:'收盘价',1:'开盘价',2:'最高价',3:'最低价',4:'成交量'},inplace=True)
    tempdf=tempdf.transpose()
    df=pd.concat([df,tempdf])
print(df.tail)
df.to_excel('白银.xlsx',encoding='utf_8_sig')
```
