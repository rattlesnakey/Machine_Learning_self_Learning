```python
from sklearn.datasets import make_blobs
#导人KNN 分类器
from sklearn.neighbors import KNeighborsClassifier
#导人画图工具
import matplotlib.pyplot as plt
#导人数据集拆分工具
from sklearn.model_selection import train_test_split
#生成样本数为200 ，分类为2 的数据集
data= make_blobs(n_samples=200,centers=2,random_state=1)
X,y = data
#将生成的数据集进行可视化
plt.scatter (X[:,0],X[:,1],c=y,cmap=plt.cm.spring,edgecolor='k')#c就是这个点的label啦
plt.show()
```

```python
print(y)
```

```python
import numpy as np
import matplotlib as mpl
elf= KNeighborsClassifier()
elf . fit (X, y )#这里的X代表的是一个二维的Point(x,y)
#下面的代码用于画图
x_min , x_max = X [:,0].min()- 1,X[:,0].max() + 1#(x,y)
y_min , y_max = X [:,1].min()- 1,X[:,1].max() + 1
xx , yy = np .meshgrid(np . arange(x_min , x_max , .02) ,
np.arange(y_min , y_max , .02))#meshgrid就是为了构造相同size的xx,yy
Z= elf.predict(np.c_[xx . ravel() , yy. ravel()])#ravel再把他们压缩成一行,也就是x自己一个，y自己一个
Z = Z.reshape(xx.shape)
cm_light = mpl.colors.ListedColormap(['#A0FFA0', '#FFA0A0'])
cm_dark = mpl.colors.ListedColormap(['g', 'r'])
plt . pcolormesh(xx , yy , Z,cmap=cm_light)#Z这时候是
plt.scatter(X[:,0],X[:, 1],c=y,cmap=cm_dark, edgecolor= 'k')
plt . xlim(xx .min(), xx . max())#设置坐标zho
plt . ylim (yy .min(), yy . max())
plt . title ("Classifier: KNN" )
plt . show()
```
